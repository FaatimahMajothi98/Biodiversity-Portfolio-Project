### Cleaning columns

 #### Step 5: Cleaning the data

As the last step revealed, the `scientific_name` column in the `species` dataset has duplicate values, and the `conservation_status` column has many missing values. These columns need to be cleaned before any analysis can be made.

For the `scientific_name` column in the `species` dataset it would make sense to drop the duplicate rows and keep only the most recent information, since the duplicate names are irrelevant to our analysis, but the conservation status is relevant. 

For the `conservation_status` column, the missing values represent the species that are not endangered. Therefore the `NaN` values can be replaced with a suitable label, in this case `'Species of Least Concern'`.

```python
# Replacing NaN values

species = species.fillna(value='Species of Least Concern')

# Checking the changes

species['conservation_status'].unique()
```

```python
# Dropping duplicates in the species dataframe

species_clean = species.drop_duplicates(subset ='scientific_name', keep ='last')

# Checking the changes 
species_clean.describe()
```

After checking to see that there are no more rows with duplicate scientific names, we can see that the count of unique common names is still less than the full count in the `common_names` column. We can investigate this by taking a closer look at the most frequent common name, `'Dicranum Moss'`, using the code below.

```python
# Filtering the species data for rows with the common name 'Dicranum Moss'

dicr_moss = species_clean[species_clean['common_names'] == 'Dicranum Moss']

dicr_moss
```

We can see that the seven occurences of species with the common name `'Dicranum Moss'` all have different scientific names. This explains the difference in unique and total counts of common names in the `common_name` column, therefore no changes need to be made to this column.

### Merging

 #### Step 6: Merging the dataframes

Now that the data has been cleaned, it is possible to merge the two dataframes on the `'scientific_name'` column. This will combine the information from the two and make it easier to graph data for visualisation.

#### Likelihood of protected status

Lastly, to answer the question of whether species in some categories are **more likely** to be endangered or protected than others, we will **analyse the `species` dataframe** from the beginning of the project, and create a a column for the protection status of each category in the dataset. This will allow us to utilise `chi-squared tests` in order to investigate if there is a significant **difference in the percentage or rate of protected species** between the categories.

```python
# Create a new column that assigns either `protected_species` or `not_protected` to each species

species_clean.loc[:, 'protected_species'] = species_clean['conservation_status'] != 'Species of Least Concern'
species_clean.head()
```

Next, we can group the data by `category` and `protected_species` so a percentage of protected species can be calculated for each category using the resulting table.

```python
# Group the data and calculate the percentage

protected_category = species_clean.groupby(['category', 'protected_species']).size().unstack().reset_index()
protected_category.rename(columns={True: 'Is Protected', False: 'Not Protected'}, inplace=True)

protected_category['Protection Rate%'] = protected_category['Is Protected'] / (protected_category['Not Protected'] + protected_category['Is Protected']) * 100
protected_category
```

Based on the percentage of each category which requires protection, we can determine that the `Mammal` category is **under the most threat**, followed by the `Bird` and `Fish` categories. The category which is **under the least threat** is the Vascular Plant category, followed closely by the Nonvascular Plant category.

##### Chi-squared testing

Now that we have the data for the proportion of species for each category which 'Is Protected' and the proportion that is 'Not Protected', we can begin to try and answer the question: "Are species in some categories **more** likely to be endangered?".
Our **null hypothesis** is that there is **no difference** between the rate of protection required for each category, while our **alternative hypothesis** is that **there is a difference** between the rate of protection required for each category.

To begin with, we will compare the `Bird` and the `Mammal` categories, as they are the ones with the highest `Protection Rate%`.

|                | Is Protected | Not Protected |
|----------------|--------------|---------------|
| Bird           | 75           | 413           |
| Mammal         | 30           | 146           |


```python
# Create contingency table and calculate

contingency_t1 = [[75, 413], 
                [11, 146]]
chi2, pval, dof, expected = chi2_contingency(contingency_t1)

print(f"P-value: {pval:.2f}\nExpected: {expected}")
```

The contingency test returned a p-value of 0.01 (1%), which is **less than the standard significance level of 0.05**. Therefore we may conclude that **there is a relationship** between category and protection status, since difference in protection status across categories is **unlikely to be due to random chance**. For this example, we can reject the null hypothesis and conclude that species in the `Mammal` category are more likely to require protection than `Bird` species.

We will now look at two more examples, comparing some of the other categories at random.
The contingency tables for both are shown below.

|                   | Is Protected | Not Protected |
|-------------------|--------------|---------------|
| Fish              | 11           | 114           |
| Amphibian         | 7            | 72            |

|                | Is Protected | Not Protected |
|----------------|--------------|---------------|
| Reptile	     | 5            | 73            |
| Vascular Plant | 46           | 4216          |

```python
# Create second contingency table and calculate

contingency_t2 = [[11, 114], 
                [7, 72]]
chi2, pval, dof, expected = chi2_contingency(contingency_t2)

print(f"P-value: {pval:.2f}\nExpected: {expected}")

# Create third contingency table and calculate

contingency_t3 = [[5, 73], 
                [46, 4216]]
chi2, pval, dof, expected = chi2_contingency(contingency_t3)

print(f"P-value: {pval:.2f}\nExpected: {expected}")
```

For the second `chi-squared` test, the p-value returned was 1.00 (100%), **far higher than the significance level** of 0.05. For this case, the null hypothesis **can not be rejected**, since the difference in protection status between the two categories is **very likely to be due to random chance**.

However, the p-value returned from the third `chi-squared` test is 0.00 (0%), which is completely **below the significance level** of 0.05. This means we can **reject the null hypothesis** entirely, as the difference in protection status between the two categories is completely **unlikely to be due to random chance**.

These results appear to coincide with the differences between the protection rate we calculated previously. For the second test (`contingency_t2`), the protection rates for the two categories are: `Amphibian`:`8.860759%` and `Fish`:`8.800000%`. Notably, the two categories share a common habitat.
For the third test, the protection rates for the two categories are: `Reptile`:`6.410256%` and `Vascular Plant`:`1.079305%`.

A **similar protection rate** seems to coincide with a very **high p-value**, while a **differring protection rate** seems to coincide with a very **low p-value**.

Lastly, we will take a look at the relationship between the rate of protection and category across all the categories by performing a chi-squared test on the contingency table below:

| category           | Is Protected | Not Protected |
|--------------------|--------------|---------------|
| Amphibian          | 7            | 72            |
| Bird               | 75           | 413           |
| Fish               | 11           | 114           |
| Mammal             | 30           | 146           |
| Nonvascular Plant  | 5            | 328           |
| Reptile            | 5            | 73            |

```python
# Create full contingency table and calculate

contingency_tf = [
    [7, 75, 11, 30, 5, 5, 46],
    [72, 413, 114, 146, 328, 73, 4216]
]

chi2, pval, dof, expected = chi2_contingency(contingency_tf)

print(f"P-value: {pval:.2f}\nExpected: {expected}")
```

Our final chi-squared test **for the whole contingency table** returns a p-value of 0.00 (0%), which is **completely below the the significance level** of 0.05. This means that overall, there is a **significant relationship** between category and protection status in the dataset, as it is **very unlikely** that the difference in rate of protection between each category is due to random chance.

We can conclude that **some species are therefore more likely to be endangered or protected than others**.
